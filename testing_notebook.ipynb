```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
images = np.load('data/raw_image_arr.npy')
labels = np.load('data/raw_labels_arr.npy')
```

```python
images.shape
```

```python
labels.shape
```

```python
num_images = 10
plt.figure(figsize=(12, 12)) 

for i in range(num_images):
    plt.subplot(1, num_images, i + 1)
    plt.imshow(images[i])
    plt.title(f"Label: {labels[i]}", fontsize=10)
    plt.axis("off")

plt.tight_layout()
plt.show()
```
